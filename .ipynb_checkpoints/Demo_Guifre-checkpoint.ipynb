# README

## Objectives
This little demo showcases the use of old2new_fdets.py and correct_baseband.py
They both rely on the baseband_library.py file.

## Recommended folder structure:

* `<mission_name>_dataset` folder must be present. The fdets are contained in this folder.
* `vex_files` folder, containing all relevant experiment files in `<vex_files>/<mission_name>/{experiment_name}.vix` must be present

## old2new_fdets.py 
This script allows you to convert old-format fdets to new-format (JUICE-like) fdets. 

**Arguments**:
filename : str
    Required. Input file(s) or wildcard pattern specifying spectra files to process.\
    Multiple files can be specified using space-separated filenames or a wildcard pattern (e.g., '*.txt').

- `--correct_baseband` : flag (optional)
    If provided, enables baseband frequency correction for the specified input files. \
    Always followed by the --experiment_name argument.\

  

- `--experiment_name`: str (optional)
    Name of the experiment to be used when correcting baseband frequencies. \
    Must be specified when --correct_baseband is used.


**Outputs**:
- Updates the input files in place (optionally with corrected baseband frequency), updating the data format.

  
## correct_baseband_fdets.py 
This script allows you to check whether the baseband present in the fdets header is the correct one, or not. \
The baseband is corrected taking into account the vex file corresponding to a given experiment.\ 
For this reason, one has to make sure to have the correct version of the experiment vex file (see below for an explanation of how the correct baseband frequency is computed)

**Arguments**:
- `filename`: One or more input files or wildcard patterns specifying spectral data files.
- `--experiment_name`: Mandatory argument specifying the experiment name for baseband corrections.

**Outputs**:
- Updates the input files in place with corrected headers, preserving the data format.

### Computing the Correct Baseband Frequency

* M. Said gave a file containing a lot of spacecraft frequencies. The file is named "spacecraft_frequencies"
* This file was converted into a dictionary in the baseband_library.py file
* Given a fdets file, a mission and related experiment, the $FREQ and/or $IF blocks (the latter only used whenever a given station is not found in the $FREQ block. This happens rarely, only for bad vex files) of the corresponding vex file are checked to relate each partecipating station to a given channel. The channel within which the spacecraft frequency (as given  in the spacecraft_frequencies file) falls is selected as a baseband channel.

### List of Available Experiments
A list of available experiments is found in the dictionary in baseband_library.py. For any other experiment you would like to probe, it needs to be added to this dictionary. This dictionary items are directly taken from the vex file corresponding to the experiment. (anyway, see example below, where I trigger an experiment_name error).

```python
import os

# Showcases old2new_fdets.py usage

fdets_file = 'vex_dataset/Fdets.vex2014.03.14.Ht.0002.r2i.txt'
fdets_file2 = 'vex_dataset/Fdets.vex2014.03.14.Ht.0005.r2i.txt'
experiment_name = 'v0314'

os.system(f'python old2new_fdets.py {fdets_file} {fdets_file2} --correct_baseband --experiment_name {experiment_name}')
```

```python
# Showcases correct_baseband_fdets.py usage (let's fake a wrong experiment)

fdets_file = 'vex_dataset//Fdets.vex2014.03.14.Ht.0006.r2i.txt' #File to convert
experiment_name = 'v014' # Experiment name (must exist in the experiment list of the baseband_library.py file)



os.system(f'python correct_baseband_fdets.py {fdets_file} --experiment_name {experiment_name}')
```
